```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Import

```python
raw_df = pd.read_csv('../../data/raw data/JPM_balance_sheet.csv')
raw_df.head()
```

## Data Cleaning

Dropped the 2020 years as they are mostly containing NaN values and formatted the dataframe to display commas every three digits.

```python
pd.options.display.float_format = '{:,.2f}'.format
filtered_df = raw_df.drop(columns ='2020-12-31')
filtered_df.head()
```

## Data Transformation

Transposed the dataframe from a long to a wide format, and set the index to the unnamed column, which was now renamed to as years.

```python
filtered_df_t = filtered_df.set_index('Unnamed: 0').T
filtered_df_t.rename_axis('Year', axis = 1, inplace=True)
filtered_df_t = filtered_df_t.sort_index(ascending = True)
filtered_df_t
```

Renamed the index (years) from YYYY-MM-DD to YYYY format.

```python
filtered_df_t.index = pd.to_datetime(filtered_df_t.index).year
filtered_df_t
```

## Column Selection

Printed the columns to easily see what columns to analyze.

```python
filtered_df_t.columns
```

## Used the Asset to Debt Ratio and Made Visualizations

Created an Asset to Debt Ratio column for visualization.

```python
filtered_df_t['Asset to Debt Ratio'] = (filtered_df_t['Total Assets'] / filtered_df_t['Total Debt'])
filtered_df_t[['Total Assets', 'Total Debt', 'Asset to Debt Ratio']]
```

```python
plt.plot(filtered_df_t.index, filtered_df_t['Asset to Debt Ratio'], marker = 'o')
plt.grid(True)
plt.xlabel('Years')
plt.ylabel('Asset to Debt Ratio')
plt.xticks(filtered_df_t.index)
plt.title('Asset to Debt Ratio from 2021 - 2024')
```
